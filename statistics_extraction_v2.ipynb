```python
#import libraries
from osgeo import gdal, osr
import glob
```

```python
#define paths
path = r'C:/.../*.tif'#path of layers to extract statistics from
TMPpath="C:/.../"#output folder

#create list of layer images
files = glob.glob(path)
print(files)
```

```python
#extract values from each layer for all forest pixels

#create an empty txt file
f=open(TMPpath+"stats.txt", "w")
f.write("mascara_bosques;sum_burned;altitud;pendiente;temperatura;viento;vapor;radiaciones;uso;EVI;distancia_poblados;distancia_vias\n")

#create one array per file
for file in files:
    if file[-19:]=="mascara_bosques.tif":
        print (file[-19:])
        #create an array from image of interest
        ds1 = gdal.Open(file)
        b1 = ds1.GetRasterBand(1)
        arrMascBosq = b1.ReadAsArray()
        rows = arrMascBosq.shape[0]
        cols = arrMascBosq.shape[1]
    elif file[-22:]=="sum_burned_mascara.tif":
        print (file[-22:])
        ds1 = gdal.Open(file)
        b1 = ds1.GetRasterBand(1)
        arrSum = b1.ReadAsArray()
    elif file[-27:]=="MDT HISP mascara bosque.tif":
        print (file[-27:])
        ds1 = gdal.Open(file)
        b1 = ds1.GetRasterBand(1)
        arrMDT = b1.ReadAsArray()
    elif file[-21:]=="pendiente_mascara.tif":
        print (file[-21:])
        ds1 = gdal.Open(file)
        b1 = ds1.GetRasterBand(1)
        arrPendiente = b1.ReadAsArray()
    elif file[-23:]=="temperatura_mascara.tif":
        print (file[-23:])
        ds1 = gdal.Open(file)
        b1 = ds1.GetRasterBand(1)
        arrTemp = b1.ReadAsArray()
    elif file[-19:]=="vientos_mascara.tif":
        print (file[-19:])
        ds1 = gdal.Open(file)
        b1 = ds1.GetRasterBand(1)
        arrViento = b1.ReadAsArray()    
    elif file[-17:]=="vapor_mascara.tif":
        print (file[-17:])
        ds1 = gdal.Open(file)
        b1 = ds1.GetRasterBand(1)
        arrVapor = b1.ReadAsArray()         
    elif file[-23:]=="radiaciones_mascara.tif":
        print (file[-23:])
        ds1 = gdal.Open(file)
        b1 = ds1.GetRasterBand(1)
        arrRadiaciones = b1.ReadAsArray()
    elif file[-15:]=="uso_mascara.tif":
        print (file[-15:])
        ds1 = gdal.Open(file)
        b1 = ds1.GetRasterBand(1)
        arrUso = b1.ReadAsArray()
    elif file[-15:]=="EVI_mascara.tif":
        print (file[-15:])
        ds1 = gdal.Open(file)
        b1 = ds1.GetRasterBand(1)
        arrEVI = b1.ReadAsArray()
    elif file[-30:]=="distancia_poblados_mascara.tif":
        print (file[-30:])
        ds1 = gdal.Open(file)
        b1 = ds1.GetRasterBand(1)
        arrPoblados= b1.ReadAsArray()     
    elif file[-26:]=="distancia_vias_mascara.tif":
        print (file[-26:])
        ds1 = gdal.Open(file)
        b1 = ds1.GetRasterBand(1)
        arrVias= b1.ReadAsArray()    

#extract values
for row in range(rows):
    for col in range(cols):
        #if pixel value is 1 (forest), extract value from other layers
        if arrMascBosq[row][col]==1:
            f.write("1;")
            for file in files:
                if file[-22:]=="sum_burned_mascara.tif":
                    f.write(str(arrSum[row][col])+";")
            for file in files:
                if file[-27:]=="MDT HISP mascara bosque.tif":
                    f.write(str(arrMDT[row][col])+";")
            for file in files:
                if file[-21:]=="pendiente_mascara.tif":
                    f.write(str(arrPendiente[row][col])+";")
            for file in files:
                if file[-23:]=="temperatura_mascara.tif":
                    f.write(str(arrTemp[row][col])+";")
            for file in files:
                if file[-19:]=="vientos_mascara.tif":
                    f.write(str(arrViento[row][col])+";")
            for file in files:
                if file[-17:]=="vapor_mascara.tif":
                    f.write(str(arrVapor[row][col])+";")
            for file in files:
                if file[-23:]=="radiaciones_mascara.tif":
                    f.write(str(arrRadiaciones[row][col])+";")         
            for file in files:
                if file[-15:]=="uso_mascara.tif":
                    f.write(str(arrUso[row][col])+";")
            for file in files:
                if file[-15:]=="EVI_mascara.tif":
                    f.write(str(arrEVI[row][col])+";")
            for file in files:
                if file[-30:]=="distancia_poblados_mascara.tif":
                    f.write(str(arrPoblados[row][col])+";")  
            for file in files:
                if file[-26:]=="distancia_vias_mascara.tif":
                    f.write(str(arrVias[row][col])+"\n")
ds1 = None 
f.close()

       
```
